1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

```python
import pandas as pd

# Load the dataset
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv"
df = pd.read_csv(url)
```

```python
df.head(3)
```

```python
df.columns
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
# Filter the data based on the specified conditions
filtered_data = df[(df['total_claim_amount'] > 1000) & (df['response'] == 'Yes')]

# Create a new DataFrame with the filtered data
new_dataframe = filtered_data.copy()

# Display the new DataFrame
new_dataframe.head()
```

2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

```python
# Filter the data for customers who have responded "Yes"
yes_responses = df[df['response'] == 'Yes']

# Calculate the average total_claim_amount for each policy type and gender
average_claim_by_policy_gender = yes_responses.groupby(['policy_type', 'gender'])['total_claim_amount'].mean().reset_index()

# Print the result
print(average_claim_by_policy_gender)
```

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
# Group the data by 'state' and count the number of unique customers
state_customer_counts = df.groupby('state')['customer'].nunique()

# Filter states with more than 500 customers
filtered_states = state_customer_counts[state_customer_counts > 500]

filtered_states
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
# Group the data by 'education' and 'gender'
grouped = df.groupby(['education', 'gender'])

# Calculate the maximum, minimum, and median customer lifetime value for each group
result = grouped['customer_lifetime_value'].agg(['max', 'min', 'median'])

result
```
