```python
dataframes = [(df1, "cleanedsales.csv"), (df2, "cleanedpurchases.csv"), (df5, "cleanedending_inventory.csv"), (df6, "cleanedbeginning_inventory.csv")]
```

```python
# Check and drop duplicates in each DataFrame
for df, filename in dataframes:
    print(f"Processing DataFrame {filename}")
    # Check for duplicates
    duplicates = df[df.duplicated(subset=['InventoryId'], keep='first')]
    if not duplicates.empty:
        print(f"Found {len(duplicates)} duplicate(s). Dropping...")
        # Drop duplicates
        df.drop_duplicates(subset=['InventoryId'], keep='first', inplace=True)
        # Save to a new CSV with specific filename
        df.to_csv(filename, index=False)
        print(f"Cleaned DataFrame saved to {filename}")
    else:
        print("No duplicates found.\n")

print("Processing completed.")
```

```python
# Convert 'SalesDate' to datetime format
df5['SalesDate'] = pd.to_datetime(df5['SalesDate'])

# Sales trends over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='SalesDate', y='SalesQuantity', data=df5, label='Sales Quantity')
sns.lineplot(x='SalesDate', y='SalesDollars', data=df5, label='Sales Dollars')
plt.title('Sales Trends Over Time')
plt.xlabel('Sales Date')
plt.ylabel('Sales Quantity / Sales Dollars')
plt.legend()
plt.show()
```

```python
## Calculate average daily sales
df5['SalesDate'] = pd.to_datetime(df5['SalesDate'])
df5['DaysSinceLastSale'] = (pd.to_datetime('now') - df5.groupby('InventoryId')['SalesDate'].transform('max')).dt.days
df5['AverageDailySales'] = df5['SalesQuantity'] / (df5['DaysSinceLastSale'] + 1)

# Calculate average lead time for procurement (assuming you have relevant columns in df2)
df2['LeadTime'] = (pd.to_datetime(df2['ReceivingDate']) - pd.to_datetime(df2['PODate'])).dt.days
average_lead_time = df2['LeadTime'].mean()

# Calculate optimal stock level
df5['OptimalStockLevel'] = df5['AverageDailySales'] * average_lead_time

# Merge with ending inventory to see the current stock level
merged_df = pd.merge(df1, df5[['InventoryId', 'OptimalStockLevel']], on='InventoryId', how='left')
```

```python
# Merge beginning and ending inventory
merged_inventory = pd.merge(df3, df1, on='InventoryId', suffixes=('_begin', '_end'))

# Calculate total inventory on hand
merged_inventory['TotalInventoryOnHand'] = merged_inventory['onHand_begin'] + merged_inventory['onHand_end']

# Save the merged inventory to a CSV file
merged_inventory.to_csv('merged_inventory.csv', index=False)

# Top 10 highest inventory
top_10_highest_inventory = merged_inventory.nlargest(10, 'TotalInventoryOnHand')
top_10_highest_inventory.to_csv('top_10_highest_inventory.csv', index=False)

# Top 10 lowest inventory
top_10_lowest_inventory = merged_inventory.nsmallest(10, 'TotalInventoryOnHand')
top_10_lowest_inventory.to_csv('top_10_lowest_inventory.csv', index=False)

# Top 10 best sellers
top_10_best_sellers = df5.nlargest(10, 'SalesQuantity')
top_10_best_sellers.to_csv('top_10_best_sellers.csv', index=False)

# Top 10 low sellers
top_10_low_sellers = df5.nsmallest(10, 'SalesQuantity')
top_10_low_sellers.to_csv('top_10_low_sellers.csv', index=False)

# Define bins and labels
bins = [-float('inf'), 1, 240, float('inf')]
labels = ['C', 'B', 'A']

# Create the 'ABC' column based on the 'SalesQuantity'
df5['ABC'] = pd.cut(df5['SalesQuantity'], bins=bins, labels=labels, include_lowest=True)

# Save the ABC Analysis to a CSV file
df5.to_csv('abc_analysis.csv', index=False)
```

```python
# Inventory Analysis Dashboard
fig, axes = plt.subplots(3, 2, figsize=(15, 10))

# Plot 1: Top 10 High Inventory
sns.barplot(x='onHand_end', y='Description_begin', data=top_10_highest_inventory, ax=axes[0, 0])
axes[0, 0].set_title('Top 10 High Inventory')

# Plot 2: Top 10 Best Sellers
sns.barplot(x='SalesQuantity', y='Description', data=top_10_best_sellers, ax=axes[0, 1])
axes[0, 1].set_title('Top 10 Best Sellers')

# Plot 3: Top 10 Low Sellers
sns.barplot(x='SalesQuantity', y='Description', data=top_10_low_sellers, ax=axes[1, 0])
axes[1, 0].set_title('Top 10 Low Sellers')

# Plot 4: ABC Analysis
sns.countplot(x='ABC', data=df5, ax=axes[1, 1])
axes[1, 1].set_title('ABC Analysis')

#Plot 5: Inventory Presence
sns.barplot(x="Description", y="onHand", data=df3.sort_values(by='onHand', ascending=False).head(10), ax=axes[2,0])
axes[2, 0].tick_params(axis='x', rotation=45)  # Rotate x-labels
axes[2, 0].set_title('Top Products by Inventory Presence')

#Plot 6: Total Inventory on Hand by City
sns.barplot(x="City", y="onHand", data=df3.sort_values(by='onHand', ascending=False).head(10), ax=axes[2, 1])
axes[2, 1].tick_params(axis='x', rotation=45)  # Rotate x-labels
axes[2, 1].set_title('Total Inventory on Hand (top 10)')

# Add padding between subplots
plt.tight_layout()

# Add a title to the entire dashboard
fig.suptitle('Inventory Analysis Dashboard', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Set the style of seaborn for better aesthetics
sns.set(style="whitegrid")

# Create subplots for each section of the analysis
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 15))

# Section i) Sales Performance

# Total Sales Quantity Over Time
sns.lineplot(x='SalesDate', y='SalesQuantity', data=df5, ax=axes[0, 0])
axes[0, 0].set_title('Total Sales Quantity Over Time')
axes[0, 0].tick_params(axis='x', rotation=45)  # Rotate x-labels

# Total Sales Revenue Over Time
sns.lineplot(x='SalesDate', y='SalesDollars', data=df5, ax=axes[0, 1])
axes[0, 1].set_title('Total Sales Revenue Over Time')
axes[0, 1].tick_params(axis='x', rotation=45)  # Rotate x-labels

# Section ii) Customer Behavior

## Distribution of Sales Quantity (using boxplot)
sns.boxplot(x='SalesQuantity', data=df5, ax=axes[1, 0])
axes[1, 0].set_title('Distribution of Sales Quantity')
axes[1, 0].set_xlim(0, 50)  # Set x-axis limits

# Distribution of Sales Price (using boxplot)
sns.boxplot(x='SalesPrice', data=df5, ax=axes[1, 1])
axes[1, 1].set_title('Distribution of Sales Price')
axes[1, 1].set_xlim(0, 50)  # Set x-axis limits


# Section iii) Seasonal Trends

# Monthly Sales Quantity
df5['SalesDate'] = pd.to_datetime(df5['SalesDate'])

df5['Month'] = df5['SalesDate'].dt.month
monthly_sales = df5.groupby('Month')['SalesQuantity'].sum().reset_index()
sns.barplot(x='Month', y='SalesQuantity', data=monthly_sales, ax=axes[2, 0])
axes[2, 0].set_title('Monthly Sales Quantity')
axes[2, 0].tick_params(axis='x', rotation=45)  # Rotate x-labels

# Monthly Sales Revenue
monthly_revenue = df5.groupby('Month')['SalesDollars'].sum().reset_index()
sns.barplot(x='Month', y='SalesDollars', data=monthly_revenue, ax=axes[2, 1])
axes[2, 1].set_title('Monthly Sales Revenue')
axes[2, 1].tick_params(axis='x', rotation=45)  # Rotate x-labels

# Adjust layout
fig.suptitle('Sales Analysis Dashboard', fontsize=16)
plt.tight_layout()
plt.show()
```
